Instructor Example 2

```
# Define Function
import math

def distance(x,y):

  dist = math.sqrt(x*x+y*y)
  return dist

# Main Code
x_var = 3
y_var = 4

dist_origin = distance(x_var, y_var)
print('The distance from the point (', x_var, ',', y_var, ') to the origin is ', dist_origin, '.')
```
