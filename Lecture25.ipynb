Instructor Example 2

```
array = [[1,12,6],[5,17,10],[10,22,15],[7,19,13]]
row1_sum = 0
row2_sum = 0
row3_sum = 0
row4_sum = 0

for i in range(0,3,1):
  row1_sum = row1_sum + array[0][i]
  row2_sum = row2_sum + array[1][i]
  row3_sum = row3_sum + array[2][i]
  row4_sum = row4_sum + array[3][i]

rows_1_2_sum = row1_sum + row2_sum
total_sum = row1_sum + row2_sum + row3_sum + row4_sum
average = total_sum/12
print('The sum of the first two rows is', rows_1_2_sum, '.')
print('The average of all values is ', average, '.')
```

Lecture 25 Participation Problem

```
riverHeight = [[8.52,6.37,5.89],[10.27,6.05,5.68],[9.19,6.19,6.04],[7.54,5.77,5.91],[8.06,6.18,6.20]]
year3_sum = 0
day2_sum = 0
for i in range(0,4,1):
  year3_sum = year3_sum + riverHeight[i][2]
for i in range(0,2,1):
  day2_sum = day2_sum + riverHeight[1][i]
year3_mean = year3_sum/5
day2_mean = day2_sum/3
print('The average height of the river in year 3 was ', year3_mean, '.')
print('The average height on day 2 was ', day2_mean, '.')

```
