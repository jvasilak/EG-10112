Problem 1

```
import math
side_1 = 10
side_2 = 15


hypotenuse = math.sqrt(side_1*side_1+side_2*side_2)
angle_1 = math.acos((side_1/hypotenuse))*(180/math.pi)
angle_2 = math.acos((side_2/hypotenuse))*(180/math.pi)
print('The hypotenuse of the triangle is ', hypotenuse,'.')
print('The two acute angles are ', angle_1, ' and ', angle_2, ' degrees.')
```

Problem 2

```
N = input('How many Fibonacci numbers would you like to see the sum of?')
N = int(N)
fibonaci = [0]*N
fibonaci[0] = 0
fibonaci[1] = 1
fibonaci_sum = 1 
for i in range(2,N,1):
  fibonaci[i] = fibonaci[i-1]+fibonaci[i-2]
  fibonaci_sum = fibonaci_sum + fibonaci[i]

print(fibonaci)
print('The sum of the first ', N, 'fiboaci numbers is ', fibonaci_sum, '.')
```

Problem 3

```
N = input('How many numbers would you like to process? ')
N = int(N)
number = [0]*N
for i in range(0,N,1):
  number[i] = input('Enter number: ')

max_val = number[0]
index = 0
for i in range(1,N,1):
  if number[i] > max_val:
    max_val = number[i]
    index = i

print('The maximum value is ', max_val, '.')
print('The value is saved in index #', index, '.')
```

Problem 4

```
def c2f(Ctemp):
  Ftemp = (Ctemp*1.8)+32
  return Ftemp

print('T(C)  T(F)')
for i in range(0,9,2):
  C = i
  F = c2f(C)
  print(C,'   ', F)
```

Problem 5

```
years = 30
interest_rate = 0.08
initial_balance = 1000
Balance = [0]*years
Balance[0] = initial_balance

for i in range(1,years,1):
  Balance[i] = Balance[i-1]+interest_rate*Balance[i-1]
Balance[years-1] = round(Balance[years-1], 2)
print('The total balance at the beginning of year ', years, ' is ', Balance[years-1],  ' dollars.')
```
