Instructor Example 1

```
g = 9.81
m = input('What is the mass in kilograms: ')
x = input('What is the change in height of the mass in meters: ')
t = input('What is the time corresponding to the change in seconds: ')

P = (g*float(m)*float(x))/float(t)
print('The average power (Watts) extracted is ' + str(P))
```

Instructor Example 2

```
num = input('Enter a number: ')
num = float(num)

if num > 0:
  print('The number is greater than 0.')
elif num == 0:
  print('The number is equal to 0.')
else:
  print('The number is less than 0.')
```

Lecture 24 Participation

```
n = input('How many numbers would you like to see averaged? ')
n = int(n)
sum = 0
for i in range(1,n+1,1):
  number = input('Enter a number: ')
  number = float(number)
  sum = sum + number
average = sum / n
print('The average of the ', n, ' numbers is ', average, '.')
```
